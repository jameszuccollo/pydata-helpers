#Python X-13

Python wrapper on X-13 to deseasonalise time series data. Takes and returns a `pandas` dataframe. Requires [X-13](https://www.census.gov/srd/www/x13as/) to be installed.

```python
from pydatafuncs import scrape
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

##Create dummy data

```python
#dummy data
df_qna = scrape.from_ONS('qna', ['BKTL', 'ABMI', 'YBGB', 'NMRK'], 'Q')
```

```python
df_qna.head()
```

##Required steps to create deseasonalised series

 - Create a `.dat` data file for each series
 - Create a `.spc` spec file for each series
 - Create a `.mta` metadata file for the batch job
 - Call the X-13 SEATS-ARIMA programme in a temporary directory
 - Read in the results and convert to `pandas` dataframe.
 

```python
%%file deseasonalise.py

import pandas as pd
import matplotlib.dates as mdates
from subprocess import Popen, PIPE
from os import path
from numpy import transpose
from string import Template


# Create .dat file from pandas dataframe
def _make_data(name, ser, freq):
    tmpdf = pd.DataFrame.from_records(transpose([ser.index.year, ser.index.month * freq / 12]))
    tmpdf['vals'] = ser.values
    tmpdf.to_csv(name + '.dat', sep='\t',header=False, index=False)


# Create .spc file from template.spc
def _make_spec(name, freq):
    subs_dict = {'dat_file' : '"' + name + '.dat"', 'period' : freq}

    with open('template.spc') as filein:
        src = Template(filein.read())

    result = src.substitute(subs_dict)
    with open(name + ".spc", "w") as spec_file:
        spec_file.write(result)
    
    
# Create .mta file
def _make_meta(name):
    with open(name + ".mta", "w") as meta_file:
        meta_file.write(name)
    
    
#Run X-13 ARIMA-SEATS
def _run_x13(name):
    x13_programme = path.normpath("C:/Program Files (x86)/winx13/x13as/x13as.exe")

    process = Popen([x13_programme, "-m", name],
                               stdout=PIPE, 
                               stderr=PIPE)
    out, err = process.communicate()
    return out, err


# Read output into pandas dataframe
def _parse(x):
    from calendar import monthrange
    
    yr = int(str(x)[:4])
    mnth = 3 * int(str(x)[-2:])
    day = monthrange(yr, mnth)[1]
    
    return pd.datetime(yr, mnth, day)


def _read_results(name):
    return pd.read_table(name + '.d11', 
                            index_col=0,
                            parse_dates=True,
                            date_parser=_parse,
                            header=None,
                            names=['date', 'seas_adj'],
                            skiprows=2)


def deseasonalise(ser, freq):
    """freq in (1, 3, 12)"""
    name = 'blurgler'
    _make_data(name, ser, freq)
    _make_spec(name, freq)
    _make_meta(name)
    output, errors = _run_x13(name)
    #print output
    return _read_results(name)
```

##Plot results

```python
i = 0
fig, ax = plt.subplots(nrows=4, figsize=(6, 15))

for name, ser in df_qna.iteritems():
    deseas = deseasonalise(ser, 4)
    deseas.sub(ser, axis=0).div(ser, axis=0).plot(ax=ax[i])
    ax[i].set_title('Seasonal component of ' + name)
    ax[i].set_ylabel('Percentage of ' + name)
    i += 1
```
